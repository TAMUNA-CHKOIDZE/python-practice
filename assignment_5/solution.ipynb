```python
class Student:
    def __init__(self, name):
        # Initialize name and an empty list for grades
        self.name = name
        # თავიდან უნდა იყოს ცარიელი ლისტი და მეთოდის გამოძახების შემდეგ უნდა ჩაემატოს
        self.grades = []

    def add_grade(self, grade):
        # Add a new grade to the grades list
        self.grades.append(grade)

    def get_average_grade(self):
        total = 0
        for grade in self.grades:
            total += grade
        average = total / len(self.grades) if self.grades else 0
        return average

    def __str__(self):
        # Return a string representation of the student
        return f"Name: {self.name}, Grades: {self.grades}, Average: {self.get_average_grade()}"


class Classroom:
    def __init__(self):
        # Initialize an empty list for students
        # თავიდან უნდა იყოს ცარიელი ლისტი და მეთოდის გამოძახების შემდეგ უნდა ჩაემატოს
        self.students = []

    def add_student(self, student):
        # Add a student to the classroom
        self.students.append(student)

    def get_top_students(self):
        # Return the top 3 students based on average grades
        # sorted() აბრუნებს ახალ სორტირებულ სიას (ან სხვა iterable-ს), არ ცვლის ორიგინალს და ვაბრუნებ კლებადობით საშუალო არითმეტიკულის მიხედვით.
        sorted_students = sorted(self.students, key=lambda student: student.get_average_grade(), reverse=True)
        # ვიღებ პირველ სამს
        return sorted_students[:3]


    def get_failed_students(self):
        # Return students with an average grade below 51
        # გავფილტრე სტუდენტები, რომელთა ქულების საშუალო არის 51-ზე ნაკლები
        filtered_students = filter(lambda student: student.get_average_grade() < 51, self.students)
        return list(filtered_students)

print("================= student 1 ==================")
student1 = Student("Gio")
student1.add_grade(56)
student1.add_grade(45)
student1.get_average_grade()
print(student1)

print("================= student 2 ==================")
student2 = Student("Nika")
student2.add_grade(80)
student2.add_grade(90)
student2.get_average_grade()
print(student2)

print("================= student 3 ==================")
student3 = Student("Ana")
student3.add_grade(100)
student3.add_grade(100)
student3.get_average_grade()
print(student3)

print("================= student 4 ==================")
student4 = Student("Luka")
student4.add_grade(55)
student4.add_grade(40)
student4.get_average_grade()
print(student4)

print("================= students in classroom ==================")
classroom = Classroom()
classroom.add_student(student1)
classroom.add_student(student2)
classroom.add_student(student3)
classroom.add_student(student4)
print("\n".join(str(student) for student in classroom.students))

print("================= Top students ==================")
top_students = classroom.get_top_students()
for student in top_students:
    print(student)

print("================= Failed students ==================")
filtered_students = classroom.get_failed_students()
for student in filtered_students:
    print(student)
```
